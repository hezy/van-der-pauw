find the sheet resistance Rs from R1 = R_AB,CD and R2 = R_BC,AD
the solution is from the equation in a paper by L. J. van der Pauw:
"A method of measuring the resistivity and hall coefficients of Lamellae of arbitrary shape", Philips Technical Review, Vol 26, 220.

$e^{-\pi R_{AB,CD}/R_s}+e^{-\pi R_{BC,DA}/R_s}=1$

so we dfine the function:

```julia
fun(Rs) = @. exp(-π * R1 / Rs) + exp(-π * R2 / Rs) - 1
```

### The resistance $R_{AB,CD}$ and $R_{BC,AD}$


```julia
R1 = 100
R2 = 50
```

### plot the fuction


```julia
using Plots
R = LinRange(0, 10 * sqrt(R1 * R2), 200)
p = plot(R, fun(R))
p = hline!([0], linestyle=:dash)
display(p)
```

### find the solution


```julia
using Roots
root = find_zero(fun, Rs)
print(root)
p = vline!([root], linestyle=:dash)
```
